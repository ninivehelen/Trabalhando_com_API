```python
import numpy as np
import pandas as pd
import tweepy as tw 
import textblob
from IPython.display import display
from textblob import TextBlob
from googletrans import Translator 
from unidecode import unidecode
import re
```

Chave para acessar API do twitter

```python
CONSUMER_KEY = ''
CONSUMER_SECRET = ''
# Access:
ACCESS_TOKEN = ''
ACCESS_SECRET = ''
```

Função para se autenticar na API do twitter

```python
def twitter_setup():
  # Authentication and access using keys:
  auth = tw.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
  auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
  # Retorna a api autenticada:
  api = tw.API(auth)
  return api
```

Variaveis para guarda resultado da análise de sentimento dos twitters

```python
negativo =0
positivo = 0
neutro = 0
total =0
```

Extraindo os twitters sobre o filme do pantera negra 2 . Extraindo 400 twitters

```python
extractor = twitter_setup()
twiters = tw.Cursor(extractor.search,q="wakanda forever",lang="en").items(400)
```

##Análise de sentimento dos twitters

```python
for tweet in twiters:
    sentiment = TextBlob(tweet.text)
     #Exibindo...
    print("[Sent.] "+str(sentiment.polarity)+"\n")
    #Estatísticas
    total += 1
    if sentiment.polarity > 0: 
        positivo += 1 
    elif sentiment.polarity < 0: 
        negativo += 1
    else:
        neutro +=1
```

Média dos sentimentos análisados. Negativos, Positivos e Neutros


```python
media_positivo = positivo/total
media_negativo = negativo/total
media_neutro = neutro/total
print(f'Porcentagem de comentários positivos % {media_positivo:.2f}')
print(f'Porcentagem de comentários negativos % {media_negativo:.2f}')
print(f'Porcentagem de comentários neutros % {media_neutro:.2f}')
```
